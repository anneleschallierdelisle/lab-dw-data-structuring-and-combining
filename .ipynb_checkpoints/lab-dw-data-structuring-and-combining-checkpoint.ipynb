## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Get the data
import pandas as pd
import numpy as np
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url3)
```

```python
# Concatenate the 3
insurance_df = pd.concat([df1,df2,df3], axis=0).reset_index(drop=True)
insurance_df
```

```python
# White spaces in column names should be replaced by _ and Column names should be in lower case
insurance_df = insurance_df.rename(columns={col: col.replace(" ","_").lower() for col in insurance_df.columns})
```

```python
# Change state column name
insurance_df = insurance_df.rename(columns={"st": "state"})
```

```python
# create a function to clean the values m, Mallle etc
def clean_gender(gender):
    if isinstance(gender, str):  # check str necessary due to NaN
        gender = gender.lower()
        if gender.startswith("m"):
            return "M"
        elif gender.startswith("f") or gender.startswith("w"):
            return "F"
        else:
            return pd.NA  # for other values
    return pd.NA  # for NaN

# create a function to map the function clean_gender to the column you want after creating a copy
def clean_df(df: pd.DataFrame):
    df2 = df.copy()
    df2["gender"] = df2["gender"].map(clean_gender)
    return df2
```

```python
# appliquer les fonctions
cleaned_df = clean_df(insurance_df)
```

```python
# remplacer les acronymes dans state
cleaned_df['state'] = cleaned_df['state'].replace({'AZ': 'Arizona', 'WA': 'Washington'})
```

```python
# education remplacer d'une autre manière 
cleaned_df['education'] = cleaned_df['education'].apply(lambda x: "Bachelor" if x == "Bachelors" else x) #attention au x!!!
```

```python
# In Customer Lifetime Value, delete the % character
cleaned_df['customer_lifetime_value'] = cleaned_df['customer_lifetime_value'].str.replace('%','', regex=False)
```

```python
# insurance_df['vehicle_class'] 
car_replacement = ["sports car", "luxury suv", "luxury car"]
cleaned_df['vehicle_class'] = cleaned_df['vehicle_class'].apply(lambda x: "luxury" if isinstance(x, str) and x.lower() in car_replacement else x)
```

```python
# customer life value corrections
cleaned_df['customer_lifetime_value'] = pd.to_numeric(cleaned_df['customer_lifetime_value'], errors='coerce')
# # of complaints corrections
cleaned_df['number_of_open_complaints'] = cleaned_df['number_of_open_complaints'].str.split('/').str[1].astype(float)
```

```python
cleaned_df['customer_lifetime_value'] = cleaned_df['customer_lifetime_value'].replace('%','') 
```

```python
numeric_cols = cleaned_df.select_dtypes(include = 'number')
```

```python
cleaned_df = cleaned_df.astype({col: float for col in numeric_cols}) # en vrai on devrait avoir int
```

```python
# help by AI get a filter number of missing values (ie columns per row, 11 is all columns)
missing_count_per_row = cleaned_df.isna().sum(axis=1)
cleaned_df = cleaned_df[missing_count_per_row < 11]
cleaned_df
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df = df.pivot_table(index='sales_channel', values=['income'], aggfunc='sum')
pivot_df =  round(pivot_df,2)
pivot_df
```

```python
# the sales channel capitalizing the highest income is Agent, followed by Branch. Thern we can observe a sharp decline for Call Center and Web sales channels. 
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Pivot the DataFrame to see the GDP based on the country and year
pivot_df = df.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='mean')
pivot_df = round(pivot_df,0)
pivot_df
```

```python
""" Customer Lifetime Value varies slightly according to gender and education level, but the differences remain small. 
Overall, high school and master's degree levels show the highest values, while doctoral degrees show the lowest for both genders. 
The differences between men and women are not very pronounced. We can notice the Customer Liefetime Value is the highest for women (and only for this education level and for women) with a high school or below level of education"""
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table.
